```python
def find_files_inside_folder(folder_name, substring=""):
    '''
    # inputs: folder_name (str), substring (optional str)
    # outputs: array of filename strings
    '''
    files = os.listdir(folder_name)
    
    if substring != "":
        found = [s for s in files if substring in s]
        if len(found) == 0: print("no matching files found")
        return np.sort(found)
    else:
        return files
```
